```python
##############psutil###########
#Cross platform library(Linux,MacOS, Windows)
#Not part of Python standard library
#Used to retrieve information from system and interface with running processes. 
#Get information in System users and sessions. 
#Montior System utilization in real time

from pprint import pprint
import psutil

# list all the running processes
psutil.process_iter()

#Iterate overall all running process 
for process in psutil.process_iter():
    try:
        #get process name and pid name
        processname=process.name()
        processID=process.pid
        print(processname,':::',processID)
    except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            pass
        

            
```

```python
#Iterate over all running processes
AllProcesses=list()
for process in psutil.process_iter():
    #Get all the details as dictionary and append dict of process details in list
    AllProcesses.append(process.as_dict(attrs=['pid','name','cpu_percent','memory_percent']))
    
pprint(AllProcesses)
print()
```
